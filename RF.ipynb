```python
import pandas as pd
dataset = pd.read_csv("insurance_pre.csv")
dataset.head()
```

```python
dataset= pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset.head()
```

```python
features=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
target=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(criterion='absolute_error',n_estimators=200,max_features='log2')
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle as pk
pk.dump(regressor,open("final_model_random_forest_reg_assign-1.sav",'wb'))
```
